```python
import numpy as np

from tensorflow.keras.models import Sequential #모델 가져오기
from tensorflow.keras.layers import Input,Dense #Dense = 뉴론을 뜻함(full connected)
from tensorflow.keras import initializers
```

```python
def gen_sequential_model():

    model = Sequential([
        Input(3, name='input_layer'), #3개의 변수로 부터 값을 받는다.입력차원=2
        Dense(128, activation='relu', name='hidden_layer1', kernel_initializer=initializers.RandomNormal(mean=0.0,stddev=0.05,seed=42)),
        Dense(1, activation='relu', name = 'output_layer', kernel_initializer=initializers.RandomNormal(mean=0.0,stddev=0.05,seed=42)) 
        ])

    model.summary()
    model.compile(optimizer='sgd', loss='mse')#free defined되있는 optimizer와 loss function돌리면 된다.

    return model;

model=gen_sequential_model()
```

```python
#y = a + bx1 + cx2^2 + dx3^3 + e
def gen_linear_regression_dataset(numofsamples=500,a=1,b=3,c=5,d=10,e=20): #train값 만들기
    np.random.seed(42) #
    x=np.random.rand(numofsamples,3) #2차원 array
    
    bias1 = a
    bias2 = e
    
    y=bias1 + b*x[:,0] + c*x[:,1]*x[:,1] + d*x[:,2]*x[:,2]*x[:,2] + bias2

    return x,y
x,y=gen_linear_regression_dataset(numofsamples=6000)
```

```python
history=model.fit(x,y,epochs=1000,verbose=2,validation_split=0.3)
```

```python
def plot_loss_curve(history):  #print하는 모듈
    
    import matplotlib.pyplot as plt
    
    plt.figure(figsize=(15,10))
    plt.plot(history.history['loss'][1:])
    plt.plot(history.history['val_loss'][1:])
    plt.title('model loss')
    plt.ylabel('loss')
    plt.xlabel('epoch')
    plt.legend(['train','test'],loc='upper right')
    plt.show()
    
    print("train loss=",history.history['loss'][-1])
    print("test loss=",history.history['val_loss'][-1])
    
plot_loss_curve(history)
```

```python
#y = a + bx1 + cx2^2 + dx3^3 + e
def predict_new_sample(model,x,a=1,b=3,c=5,d=10,e=20):
    x=x.reshape(1,3)
    y_pred = model.predict(x)[0][0]
    
    y_actual = a + b*x[0][0] + c*x[0][1]*x[0][1] + d*x[0][2]*x[0][2]*x[0][2] + e
    print("y actual value = ",y_actual)
    print("y predicted value = ",y_pred)
    
    
x1=float(input("x1 : "))
x2=float(input("x2 ; "))
x3=float(input("x3 : "))

predict_new_sample(model,np.array([x1,x2,x3]))
```
