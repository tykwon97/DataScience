```python
import numpy as np

from tensorflow.keras.models import Sequential #모델 가져오기
from tensorflow.keras.layers import Input,Dense #Dense = 뉴론을 뜻함(full connected)
```

```python
def gen_sequential_model():

    model = Sequential([
        Input(2, name='input_layer'), #2개의 변수로 부터 값을 받는다.입력차원=2
        Dense(16, activation='sigmoid', name='hidden_layer1'),#16개의 히든 레이어에 입력
        #2(input)*16(hidden layer 뉴론)=32 + 16(bias값) = 48개
        Dense(1, activation='relu', name = 'output_layer') # 하나의 y값을 받는다.
        #16개의 weight값 + bias 1개 = 17개
        ])

    model.summary()

    model.compile(optimizer='sgd', loss='mse') #free defined되있는 optimizer와 loss function돌리면 된다.

    #linear regression -> mse
    #multiclass classification, binary classification -> cross entropy

    return model;
```

```python
#y = w1*x1 + w2*x2 + b
def gen_linear_regression_dataset(numofsamples=500,w1=3,w2=5,b=10): #train값 만들기
    np.random.seed(0) #
    x=np.random.rand(numofsamples,2) #2차원 array
    print(x)
    print(x.shape)
    
    coef = np.array([w1,w2]) #행렬 이용해서 한번에 계산
    bias = b
    
    print(coef)
    print(coef.shape)
    
    y=np.matmul(x, coef.transpose()+bias) #matrix multipication
    print(y)
    print(y.shape)
    
    
gen_linear_regression_dataset(numofsamples=1000)
```
