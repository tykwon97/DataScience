```python
import numpy as np
import pandas as pd

xlfile = "C:\\2020\\데이터사이언스개론\\과제 #2(Python 과제)\db_score.xlsx"
df = pd.read_excel(xlfile)
del df['sno']

df_grade = pd.read_excel(xlfile) #grade가 없는 dataframe
del df_grade['sno']
del df_grade['grade']

#median (평균값) -----------------------------------------------
print("<mean (평균값)>")
print(df_grade.mean(),'\n')
#median (중간값) -----------------------------------------------
print("<median (중간값)>")
print(df_grade.median(),'\n')
#mode-----------------------------------------------------------
print("<mode>")
print(df_grade.mode(),'\n') #score은 겹치는 수가 없어서 모든 값이 출력된다.
#variance (분산) -----------------------------------------------
print("<variance (분산)>")
print(df_grade.var(),'\n')
#standard deviation (표준편차) ---------------------------------
print("<standard deviation (표준편차)>")
print(df_grade.std(),'\n')
```

```python
average=[0,0,0,0,0,0] #평균값 저장 변수

average[0]=df_grade['attendance'].mean()
average[1]=df_grade['homework'].mean()
average[2]=df_grade['discussion'].mean()
average[3]=df_grade['midterm'].mean()
average[4]=df_grade['final'].mean()
average[5]=df_grade['score'].mean()

average_df=abs(df_grade-average)

#AAD------------------------------------------------
print("<AAD>\n")

print("attendance : ", average_df.attendance.mean())
print("homework : ", average_df.homework.mean())
print("discussion : ", average_df.discussion.mean())
print("midterm : ", average_df.midterm.mean())
print("final : ", average_df.final.mean())
print("score : ", average_df.score.mean(),'\n')

#MAD------------------------------------------------
print("<MAD-과제의 M(median)AD>\n")

print("attendance : ", average_df.attendance.median())
print("homework : ", average_df.homework.median())
print("discussion : ", average_df.discussion.median())
print("midterm : ", average_df.midterm.median())
print("final : ", average_df.final.median())
print("score : ", average_df.score.median(),'\n')

print("<MAD-내장함수를 이용한 M(mean)AD>")
print("-AAD와 같은 값이 나온다.\n")
print(df_grade.mad(),'\n')
```
